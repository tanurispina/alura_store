### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python

# Calculando o faturamento total de cada loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Exibindo os resultados formatados na tela
print(f"Faturamento da Loja 1: R$ {faturamento_loja1:,.2f}")
print(f"Faturamento da Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento da Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento da Loja 4: R$ {faturamento_loja4:,.2f}")
```

# 2. Vendas por Categoria

```python
# Contando a quantidade de produtos vendidos por categoria em cada loja
vendas_categoria_loja1 = loja['Categoria do Produto'].value_counts()
vendas_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
vendas_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
vendas_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

# Exibindo os rankings na tela
print("Categorias Mais Vendidas: Loja 1")
print(vendas_categoria_loja1)
print("\nCategorias Mais Vendidas: Loja 2")
print(vendas_categoria_loja2)
print("\nCategorias Mais Vendidas: Loja 3")
print(vendas_categoria_loja3)
print("\nCategorias Mais Vendidas: Loja 4")
print(vendas_categoria_loja4)
```

# 3. Média de Avaliação das Lojas


```python
media_loja1 = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média da Loja 1: {media_loja1:.2f}")
print(f"Média da Loja 2: {media_loja2:.2f}")
print(f"Média da Loja 3: {media_loja3:.2f}")
print(f"Média da Loja 4: {media_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
# Contando as vendas por produto em cada loja
produtos_loja1 = loja['Produto'].value_counts()
produtos_loja2 = loja2['Produto'].value_counts()
produtos_loja3 = loja3['Produto'].value_counts()
produtos_loja4 = loja4['Produto'].value_counts()

# Exibindo os mais e menos vendidos da Loja 1
print("Loja 1")
print("Mais vendidos:\n", produtos_loja1.head(3))
print("Menos vendidos:\n", produtos_loja1.tail(3))
print("\n")

# Exibindo os mais e menos vendidos da Loja 2
print("Loja 2")
print("Mais vendidos:\n", produtos_loja2.head(3))
print("Menos vendidos:\n", produtos_loja2.tail(3))
print("\n")

# Exibindo os mais e menos vendidos da Loja 3
print("Loja 3")
print("Mais vendidos:\n", produtos_loja3.head(3))
print("Menos vendidos:\n", produtos_loja3.tail(3))
print("\n")

# Exibindo os mais e menos vendidos da Loja 4
print("Loja 4")
print("Mais vendidos:\n", produtos_loja4.head(3))
print("Menos vendidos:\n", produtos_loja4.tail(3))
```

# 5. Frete Médio por Loja

```python
frete_loja1 = loja['Frete'].mean()
frete_loja2 = loja2['Frete'].mean()
frete_loja3 = loja3['Frete'].mean()
frete_loja4 = loja4['Frete'].mean()

print(f"Frete Médio da Loja 1: R$ {frete_loja1:.2f}")
print(f"Frete Médio da Loja 2: R$ {frete_loja2:.2f}")
print(f"Frete Médio da Loja 3: R$ {frete_loja3:.2f}")
print(f"Frete Médio da Loja 4: R$ {frete_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Preparando os dados
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [loja['Preço'].sum(), loja2['Preço'].sum(), loja3['Preço'].sum(), loja4['Preço'].sum()]

# Criando o Gráfico
plt.figure(figsize=(8, 5))
plt.bar(lojas, faturamentos, color=['#4CAF50', '#2196F3', '#FFC107', '#F44336'])
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.ticklabel_format(style='plain', axis='y') # Evita número científico
plt.show()
```

```python
# Preparando os dados
avaliacoes = [loja['Avaliação da compra'].mean(), loja2['Avaliação da compra'].mean(), loja3['Avaliação da compra'].mean(), loja4['Avaliação da compra'].mean()]

# Criando o Gráfico
plt.figure(figsize=(8, 5))
plt.plot(lojas, avaliacoes, marker='o', color='purple', linestyle='dashed', linewidth=2, markersize=8)
plt.title('Média de Avaliação dos Clientes por Loja')
plt.xlabel('Lojas')
plt.ylabel('Nota Média (1 a 5)')
plt.ylim(0, 5) # Fixando a nota máxima em 5
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Pegando as 5 categorias mais vendidas da Loja 1
top_categorias = loja['Categoria do Produto'].value_counts().head(5).sort_values()

# Criando o Gráfico
plt.figure(figsize=(8, 5))
top_categorias.plot(kind='barh', color='teal')
plt.title('Top 5 Categorias Mais Vendidas (Loja 1)')
plt.xlabel('Quantidade de Produtos Vendidos')
plt.ylabel('Categoria')
plt.show()
```

```python
# Criando o Gráfico com dados da Loja 1
plt.figure(figsize=(8, 5))
plt.scatter(loja['Preço'], loja['Frete'], color='darkorange', alpha=0.6, edgecolors='black')
plt.title('Relação entre Preço do Produto e Valor do Frete (Loja 1)')
plt.xlabel('Preço do Produto (R$)')
plt.ylabel('Valor do Frete (R$)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Criando o Histograma com os preços da Loja 1
plt.figure(figsize=(8, 5))
plt.hist(loja['Preço'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribuição dos Preços dos Produtos (Loja 1)')
plt.xlabel('Faixa de Preço (R$)')
plt.ylabel('Frequência (Quantidade de Vendas)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Contando as vendas por tipo de pagamento na Loja 1
tipos_pagamento = loja['Tipo de pagamento'].value_counts()

# Criando o Gráfico de Pizza
plt.figure(figsize=(8, 5))
plt.pie(tipos_pagamento, labels=tipos_pagamento.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0'])
plt.title('Preferência de Tipo de Pagamento (Loja 1)')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

# Plotando os pontos de cada loja (usando as colunas 'lat' e 'lon')
# OBS: Se der erro de coluna não encontrada, tente mudar 'lon' para 'Longitude' e 'lat' para 'Latitude'
plt.scatter(loja['lon'], loja['lat'], color='#4CAF50', label='Loja 1', alpha=0.5, s=20)
plt.scatter(loja2['lon'], loja2['lat'], color='#2196F3', label='Loja 2', alpha=0.5, s=20)
plt.scatter(loja3['lon'], loja3['lat'], color='#FFC107', label='Loja 3', alpha=0.5, s=20)
plt.scatter(loja4['lon'], loja4['lat'], color='#F44336', label='Loja 4', alpha=0.5, s=20)

plt.title('Distribuição Geográfica das Vendas - Todas as Lojas')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
import folium

# Pegando a média das coordenadas para centralizar o mapa
centro_lat = loja['lat'].mean()
centro_lon = loja['lon'].mean()

# Criando o mapa base
mapa_vendas = folium.Map(location=[centro_lat, centro_lon], zoom_start=5)

# Adicionando um marcador para as 100 primeiras vendas da Loja 1
for indice, linha in loja.head(100).iterrows():
    folium.CircleMarker(
        location=[linha['lat'], linha['lon']],
        radius=5,
        color='green',
        fill=True,
        fill_color='green',
        fill_opacity=0.6,
        popup=f"Produto: {linha['Produto']} | R$ {linha['Preço']}"
    ).add_to(mapa_vendas)

# Exibindo o mapa (no Colab é só chamar a variável no final)
mapa_vendas
```
